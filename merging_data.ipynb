# Merging of four datasets into one

```python
# Import of library in order to concatmerge all the datasets
import pandas as pd
```

```python
# Reading the csv files and concatenation. Show first 5 rows of the merged dataset
files = ['opel_corsa_01.csv','opel_corsa_02.csv','peugeot_207_01.csv','peugeot_207_02.csv']
dfs = [pd.read_csv(f) for f in files]
df = pd.concat(dfs, ignore_index=True)
df.head()
```

```python
df = df.drop(df.columns[0],axis=1) # unnamed column drop
df.head()
```

```python
df.info() # general info about the dataset
```

```python
df = df.to_csv("data.csv",index=False) # saving dataset into data.csv file
```
